```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
import scipy.stats as stats
%matplotlib inline
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv("train.csv")
df.info()
```

```python
df.head()
```

```python
for i in df.columns:
    if df[i].isna().sum()!=0:
        print(i, df[i].isna().sum())
df.describe()
```

```python
df = df.drop(["PassengerId","Cabin", "Name", "Ticket"], axis=1)
```

```python
df.head()
```

```python
y = df.Survived
```

```python
enc = LabelEncoder()
enc.fit_transform(list(df.Sex.unique()))
```

```python
df.Sex = enc.transform(df.Sex)
df.head()
```

```python
df = df.fillna({'Age':df.Age.mean()})
df.describe()
```

```python
for i in df.columns:
    if df[i].isna().sum()!=0:
        print(i, df[i].isna().sum())
df.describe()
```

```python
enc2 = LabelEncoder()
enc2.fit_transform(df.Embarked)
df.Embarked = enc2.transform(df.Embarked)
```

```python
for i in df.columns:
    if df[i].isna().sum()!=0:
        print(i, df[i].isna().sum())
df.describe()
```

```python
corr = df.corr()
corr
```

From this we can see that most of the columns of the matrix are not correlated with one another so they are not showing much similarities so all columns can be considered

```python
X = df.drop(['Survived', 'Age', "SibSp", "Parch"], axis = 1)
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
sns.heatmap(df.corr(), linewidths=5, square= False, annot = True)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['experimental_proprty'] = y.values
plt.figure(figsize=(10, 8))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='experimental_proprty', palette='viridis', s=100, alpha=0.7)
plt.title('PCA of Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Experimental Property')
plt.show()
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
clf = SGDClassifier(loss = "hinge", penalty = "l2", max_iter = 30)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
test = pd.read_csv("test.csv")
for i in test.columns:
    if test[i].isna().sum()!=0:
        print(i, test[i].isna().sum())
test.describe()
```

```python
passid = test.PassengerId
```

```python
test = test.drop(['Age', "SibSp", "Parch", "PassengerId","Cabin", "Name", "Ticket"], axis = 1)
```

```python
test.head()
```

```python
test = test.fillna({"Fare":test.Fare.mean()})
```

```python
for i in test.columns:
    if test[i].isna().sum()!=0:
        print(i, test[i].isna().sum())
test.describe()
```

```python
test.head()
```

```python
test.Sex = enc.transform(test.Sex)
test.Embarked = enc2.transform(test.Embarked)
```

```python
test.describe()
```

```python
X_scaled = scaler.transform(test)
y_predicted = clf.predict(X_scaled)
```

```python
y_predicted
```

```python
output = pd.DataFrame({'PassengerId': passid, 'Survived': y_predicted})
```

```python
output.to_csv("Submission.csv", index = False)
```
